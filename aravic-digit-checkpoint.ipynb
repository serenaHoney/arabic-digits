```python

import numpy as np
import librosa 
import matplotlib.pyplot as plt
import IPython.display as ipd
import scipy


 
arabic_digit = r"C:\Users\elizh\OneDrive\Documents\Academics\University Maths\Year 2\Arabic_digit_project\Train_Arabic_Digit.txt"

data = np.loadtxt(arabic_digit)
print(data)
mfcc = data[2:39]
wav = librosa.feature.inverse.mfcc_to_audio(mfcc)
print('wav: ', wav)
def show_data(data):
    plt.figure()
    plt.title("data")
    plt.plot(data)
    plt.show()

show_data(wav)
ipd.Audio(wav, rate=11025)
```
